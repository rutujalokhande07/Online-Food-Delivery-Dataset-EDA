**Online Food Delivery Dataset - Exploratory Data Analysis (EDA)**

```python
#Import the Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import Kagglehub
import kagglehub

# Download dataset
path = kagglehub.dataset_download("sudarshan24byte/online-food-dataset")

print("Path to dataset files:", path)



```

```python
#Load the File
df = pd.read_csv(f"{path}/onlinefoods.csv")
df
```

```python
#Q.1 Load the dataset and display the first 10 records.
df.head(10)
```

```python
#Q.2 Find the total number of rows and columns in the dataset.
df.shape
```

```python
#.3 Identify the data type of each feature.
df.dtypes
```

```python
#Q.4. Check whether the dataset contains any missing values.
df.isnull().sum()
```

```python
#Q.5. Generate the statistical summary of all numerical columns.
df.describe()
```

```python
#Q.6. List all categorical and numerical features separately.
categorical_features = df.select_dtypes(include=['object']).columns
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns

print("Categorical Features:")
print(categorical_features)

print("\nNumerical Features:")
print(numerical_features)
```

```python
#Q.7. Plot the distribution of customer ages using a histogram.
plt.hist(df['Age'], bins=20, color='green', edgecolor='black')
plt.title('Distribution of Customer Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

```python
#Q.8. Create a count plot showing the number of customers in each occupation.
sns.countplot(data=df, x='Occupation',hue='Occupation', palette='Set2',legend=False)
plt.title('Number of Customers in Each Occupation')
plt.show()
```

```python
#Q.9. Plot the distribution of monthly income categories.
sns.countplot(x='Monthly Income',hue='Monthly Income',data=df,palette="Set3",legend=False)
plt.title('Distribution of Monthly Income Categories')
plt.xlabel('Monthly Income')
plt.ylabel('Number of Customers')
plt.xticks(rotation=10)
plt.show()
```

```python
#Q.10. Find the top 5 most common educational qualifications.
df['Educational Qualifications'].value_counts().head(5)
```

```python
#Q.11. Compare the distribution of Gender across different Occupations using an appropriate visualization.
sns.countplot(data=df, x='Occupation',hue='Gender', palette='Set3')
plt.title('Distribution of Gender Across Different Occupations')
```

```python
#Q.12. Analyze the relationship between Marital Status and Feedback.
sns.countplot(data=df, x='Marital Status', hue='Feedback',palette=['red', 'yellow'])

plt.title('Relationship between Marital Status and Feedback')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.grid(axis='y', alpha=0.3)
plt.legend(title='Feedback')
```

```python
#Q.13. Compare Family Size across different Monthly Income groups using a box plot.
sns.boxplot(data=df,x='Monthly Income',y='Family size',hue='Monthly Income',palette='Set2',legend=False)

plt.title('Compare Family Size across Different Monthly Income Groups')
plt.xticks(rotation=20)
```

```python
#Q.14. Visualize the relationship between Age and Family Size using a scatter plot.
sns.scatterplot(data=df,x='Age',y='Family size',hue='Family size',palette='Set2')

plt.title('Relationship between Age and Family Size')
plt.xlabel('Age')
plt.ylabel('Family Size')
plt.legend(title="Family size", loc="upper right")
plt.grid(alpha=0.3)
```

```python
#Q.15. Create a heatmap of correlations among all numerical features.
numeric_features = df.select_dtypes(include=['int64', 'float64'])

correlation_matrix = numeric_features.corr()

plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='Purples', fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap of Numerical Features")
```

```python
#Q.16. Analyze how Occupation, Monthly Income, and Feedback are related.

df.groupby(['Occupation', 'Monthly Income', 'Feedback']).size().reset_index(name='Count')
```

```python
sns.countplot(data=df,x='Occupation',hue='Feedback',palette='Set2')

plt.title("Relationship among Occupation and Feedback")
plt.xlabel("Occupation")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.grid(axis='y', alpha=0.3)
```

```python
#Q.17. Compare the average family size for each educational qualification.
df.groupby('Educational Qualifications')['Family size'].mean()
```

```python
#Q.18. Plot customer locations using Latitude and Longitude on a scatter plot and identify any visible clusters.
sns.scatterplot(data=df,x='longitude',y='latitude',color='purple',s=60)

plt.title('Customer Locations using Latitude and Longitude')

plt.grid(alpha=0.3)
```

```python
#Q.19. Determine which Pin Codes have the highest number of customers.
df['Pin code'].value_counts().head(5)
```
